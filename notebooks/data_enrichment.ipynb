#### Append to Existing Data

```python
# Append new records
main_updated = pd.concat([main, new_observations, new_events], ignore_index=True)
impacts_updated = pd.concat([impacts, new_impacts], ignore_index=True)

print("Updated main shape:", main_updated.shape)
print("Updated impacts shape:", impacts_updated.shape)
```

### Validations

```python
# Validate parent_ids in the updated dataset
missing_parents = set(impacts_updated['parent_id']) - set(main_updated['record_id'])
if missing_parents:
    print("Warning: These parent_ids are missing in main_updated:", missing_parents)
else:
    print("All parent_ids are valid in main_updated")
```

Check for Duplicate Record IDs

```python
duplicates = main_updated['record_id'][main_updated['record_id'].duplicated()]
if not duplicates.empty:
    print("Warning: Duplicate record_ids found in main_updated:", duplicates.tolist())
```
